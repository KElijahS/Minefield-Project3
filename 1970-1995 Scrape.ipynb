```python
import pandas as pd
```

# Data for 70s top 100

```python
url_70s = 'http://billboardtop100of.com/1970-2/'
```

```python
data_70s = pd.read_html(url_70s)
```

```python
data_70s[0]
```

```python
df_70s = data_70s[0]
pd.DataFrame.to_csv(df_70s, 'top70s.csv')
```
